```python
#Session 9 assignment by Sreedhara Jagatagar Sreenivasa
#Read the dataset from the below link
#https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

#Import Pandas Libraris
import pandas as pd

#Read the file and store in a Data Frame
baby_df_ld=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

```python
#print first 10 rows
baby_df=baby_df_ld
baby_df_ld.head(10)
```

```python
#Session 9 assignment by Sreedhara Jagatagar Sreenivasa
#1. Delete unnamed columns

#Drop field which is not named
baby_df.drop(labels='Unnamed: 0',axis=1,inplace=True)

#Print first 10 rows
baby_df.head(10)
```

```python
#Session 9 assignment by Sreedhara Jagatagar Sreenivasa
#2. Show the distribution of male and female

#Distribute Male and Female using Group functions.
baby_df.groupby(['Gender']).count()
```

```python
#Session 9 assignment by Sreedhara Jagatagar Sreenivasa
#3. Show the top 5 most preferred names

#Display 5 most preferred names. Sort in ascending orders by count and select last 5 rows which gives most preffered or Used Names as count is highest.
baby_df.sort_values(['Count']).tail(5)
```

```python
#Session 9 assignment by Sreedhara Jagatagar Sreenivasa
#4. What is the median name occurence in the dataset

#Calculate and print Median Name Occurance
print('Median name Occurence is: ' ,baby_df['Count'].median())
print(' ')
#Calculate Distance Median and create new filed as Dist
baby_df['Dist']=abs(baby_df['Count']-baby_df['Count'].median())
print('Distance from Median with Dist Field as below:')
print(' ')
print(baby_df.head(5))
print(' ')
print('Distance Median is: ' ,baby_df['Dist'].median())
```
